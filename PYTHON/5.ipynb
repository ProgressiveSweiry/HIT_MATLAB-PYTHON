```python
import numpy as np

def PearsonCorrelation(xrr,yrr):
  sum1 = 0
  sum2 = 0
  sum3 = 0
  x_mean = np.mean(xrr)
  y_mean = np.mean(yrr)
  for i in range(len(xrr)):
    sum1 += (xrr[i] - x_mean)*(yrr[i] - y_mean)
    sum2 += (xrr[i] - x_mean)**2
    sum3 += (yrr[i] - y_mean)**2
  return (sum1 / np.sqrt(sum2 * sum3))
```

```python
X = np.array([9,7,5,3,1])
Y = np.array([10,6,1,5,3])
R1 = PearsonCorrelation(X,Y)
print('R1: {}\n'.format(R1))

X = np.array([1,0,1])
Y = np.array([0,2,1])
R2 = PearsonCorrelation(X,Y)
print('R2: {}\n'.format(R2))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def PlotFun(myStr, par):
  t = np.linspace(par[0], par[1], par[2])
  y1 = 10 * np.sin(t)
  y2 = 5 * np.cos(3*t)
  plt.plot(t,y1,'b',t,y2,'r')
  plt.title(myStr)
```

```python
PlotFun('MyTitle', [0,10,1000])
```
